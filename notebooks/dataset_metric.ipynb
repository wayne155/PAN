```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, ccf

# 绘制自相关图
cross_corr = ccf(data[:, 0], data[:, 1])
plt.figure(figsize=(10, 5))
plt.stem(cross_corr)
plt.title('Cross-Correlation between var1 and var2')
plt.xlabel('Lag')
plt.ylabel('Cross-Correlation')
plt.show()

# 显示图形
plt.show()
```
