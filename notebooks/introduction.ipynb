```python
import numpy as np
import matplotlib.pyplot as plt

# 原始直线上升趋势
x = np.linspace(0, 100, 100)  # 时间轴
y = np.linspace(0, 100, 100)  # 上升趋势值

# 分成若干patch
patch_size = 20
patches = [y[i:i + patch_size] for i in range(0, len(y), patch_size)]

# 正向打乱patch顺序
np.random.seed(42)  # 为了可复现结果
shuffled_patches = np.random.permutation(patches)
y_shuffled = np.concatenate(shuffled_patches)

# 反转patch顺序
reversed_patches = patches[::-1]
y_reversed = np.concatenate(reversed_patches)

# 绘制图像
plt.figure(figsize=(12, 6))

# 原始直线
plt.subplot(1, 3, 1)
plt.plot(x, y, label='Original Line', color='blue')
plt.title('Original Line')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)

# 正向打乱patch顺序
plt.subplot(1, 3, 2)
plt.plot(x, y_shuffled, label='Shuffled Patches', color='orange')
plt.title('Shuffled Patches (Order)')
plt.xlabel('Time')
plt.grid(True)

# 反转patch顺序
plt.subplot(1, 3, 3)
plt.plot(x, y_reversed, label='Reversed Patches', color='green')
plt.title('Reversed Patches (Order)')
plt.xlabel('Time')
plt.grid(True)

plt.tight_layout()
plt.show()
```
