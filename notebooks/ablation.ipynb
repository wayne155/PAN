```python
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots

# Data preparation
datasets = ["ETTh1", "ETTh2", "ETTm1", "ETTm2", "Weather", "Electricity", "PEMS08", "PEMS-BAY"]
methods = ["PAN", "w/o ci", "w/o patch", "w/o al"]
# plt.style.use(['science', 'ieee', 'notebook'])
plt.style.use(['science', 'notebook'])

# Mean values
mean_values = [
    [0.558, 0.565, 0.710, 0.566],
    [0.289, 0.294, 0.330, 0.300],
    [0.466, 0.479, 0.564, 0.482],
    [0.264, 0.270, 0.293, 0.270],
    [0.298, 0.303, 0.308, 0.299],
    [0.295, 0.299, 0.346, 0.308],
    [0.173, 0.188, 0.207, 0.174],
    [0.259, 0.272, 0.336, 0.260]
]

# Variance values
var_values = [
    [0.007808724, 0.005784088, 0.018835514, 0.016361251],
    [0.009555771, 0.009881566, 0.009882574, 0.003474532],
    [0.002283078, 0.002996322, 0.006486433, 0.003974036],
    [0.001403831, 0.001242603, 0.001862755, 0.002993283],
    [0.001324592, 0.001026326, 0.002797566, 0.00149117],
    [0.000721306, 0.000926144, 0.002429069, 0.001896462],
    [0.000455446, 0.00042624, 0.000981254, 0.000382158],
    [0.000480388, 0.000546326, 0.001717888, 0.000397103]
]

# Create a DataFrame for processing
data = []
for i, dataset in enumerate(datasets):
    for j, method in enumerate(methods):
        data.append([dataset, method, mean_values[i][j], var_values[i][j]])

df = pd.DataFrame(data, columns=["Dataset", "Method", "Mean", "Variance"])

# Plot candle chart for each dataset
for dataset in datasets:
    subset = df[df["Dataset"] == dataset]
    
    plt.figure(figsize=(8, 6))
    plt.errorbar(
        subset["Method"],
        subset["Mean"],
        yerr=subset["Variance"],
        fmt='o',
        capsize=5,
        label="Mean ± Variance"
    )
    plt.title(f"Dataset: {dataset}")
    plt.xlabel("Method")
    plt.ylabel("Performance")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
