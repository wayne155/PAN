```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 
np.random.seed(42)
days = 365  # 365
seasons = ['Spring', 'Summer', 'Autumn', 'Winter']  # 
season_days = [90, 90, 90, 95]  # 

# 
seasonal_cycle = []
for season, days_in_season in zip(seasons, season_days):
    seasonal_cycle += [season] * days_in_season
seasonal_cycle = np.array(seasonal_cycle)

# 
temperature = np.zeros(days)
for i in range(days):
    if seasonal_cycle[i] == 'Spring':
        temperature[i] = 15 + np.random.normal(0, 2)  # 15
    elif seasonal_cycle[i] == 'Summer':
        temperature[i] = 30 + np.random.normal(0, 3)  # 30
    elif seasonal_cycle[i] == 'Autumn':
        temperature[i] = 20 + np.random.normal(0, 2)  # 20
    elif seasonal_cycle[i] == 'Winter':
        temperature[i] = 5 + np.random.normal(0, 4)  # 5

# 
electricity = np.zeros(days)
for i in range(days):
    if seasonal_cycle[i] == 'Spring':
        electricity[i] = 50 + np.random.normal(0, 5)  # 
    elif seasonal_cycle[i] == 'Summer':
        electricity[i] = 80 + 2 * (temperature[i] - 25) + np.random.normal(0, 8)  # 
    elif seasonal_cycle[i] == 'Autumn':
        electricity[i] = 60 + np.random.normal(0, 5)  # 
    elif seasonal_cycle[i] == 'Winter':
        electricity[i] = 90 - 1.5 * temperature[i] + np.random.normal(0, 10)  # 

# DataFrame
data = pd.DataFrame({
    'Day': np.arange(days),
    'Season': seasonal_cycle,
    'Temperature': temperature,
    'Electricity': electricity
})

# 
plt.figure(figsize=(14, 6))

# 
plt.subplot(2, 1, 1)
plt.plot(data['Day'], data['Temperature'], label='Temperature', color='orange')
plt.title('Temperature Over the Year')
plt.xlabel('Day')
plt.ylabel('Temperature (C)')
plt.legend()

# 
plt.subplot(2, 1, 2)
plt.plot(data['Day'], data['Electricity'], label='Electricity Usage', color='blue')
plt.title('Electricity Usage Over the Year')
plt.xlabel('Day')
plt.ylabel('Electricity (kWh)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 
def calculate_rolling_correlation(x, y, window):
    rolling_corr = []
    for i in range(len(x) - window + 1):
        corr = np.corrcoef(x[i:i+window], y[i:i+window])[0, 1]
        rolling_corr.append(corr)
    return np.array(rolling_corr)
```
