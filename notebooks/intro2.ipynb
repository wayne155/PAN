```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 设置参数
np.random.seed(42)
days = 365  # 一年365天
seasons = ['Spring', 'Summer', 'Autumn', 'Winter']  # 四季
season_days = [90, 90, 90, 95]  # 每个季节的天数

# 生成季节标签
seasonal_cycle = []
for season, days_in_season in zip(seasons, season_days):
    seasonal_cycle += [season] * days_in_season
seasonal_cycle = np.array(seasonal_cycle)

# 生成气温数据
temperature = np.zeros(days)
for i in range(days):
    if seasonal_cycle[i] == 'Spring':
        temperature[i] = 15 + np.random.normal(0, 2)  # 春天温度：平均15度
    elif seasonal_cycle[i] == 'Summer':
        temperature[i] = 30 + np.random.normal(0, 3)  # 夏天温度：平均30度
    elif seasonal_cycle[i] == 'Autumn':
        temperature[i] = 20 + np.random.normal(0, 2)  # 秋天温度：平均20度
    elif seasonal_cycle[i] == 'Winter':
        temperature[i] = 5 + np.random.normal(0, 4)  # 冬天温度：平均5度

# 根据气温生成电量数据
electricity = np.zeros(days)
for i in range(days):
    if seasonal_cycle[i] == 'Spring':
        electricity[i] = 50 + np.random.normal(0, 5)  # 春季电量低
    elif seasonal_cycle[i] == 'Summer':
        electricity[i] = 80 + 2 * (temperature[i] - 25) + np.random.normal(0, 8)  # 夏季电量高，随温度升高
    elif seasonal_cycle[i] == 'Autumn':
        electricity[i] = 60 + np.random.normal(0, 5)  # 秋季电量中等
    elif seasonal_cycle[i] == 'Winter':
        electricity[i] = 90 - 1.5 * temperature[i] + np.random.normal(0, 10)  # 冬季电量高，随温度降低

# 转为DataFrame
data = pd.DataFrame({
    'Day': np.arange(days),
    'Season': seasonal_cycle,
    'Temperature': temperature,
    'Electricity': electricity
})

# 可视化
plt.figure(figsize=(14, 6))

# 气温子图
plt.subplot(2, 1, 1)
plt.plot(data['Day'], data['Temperature'], label='Temperature', color='orange')
plt.title('Temperature Over the Year')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')
plt.legend()

# 电量子图
plt.subplot(2, 1, 2)
plt.plot(data['Day'], data['Electricity'], label='Electricity Usage', color='blue')
plt.title('Electricity Usage Over the Year')
plt.xlabel('Day')
plt.ylabel('Electricity (kWh)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 滑动相关性计算函数
def calculate_rolling_correlation(x, y, window):
    rolling_corr = []
    for i in range(len(x) - window + 1):
        corr = np.corrcoef(x[i:i+window], y[i:i+window])[0, 1]
        rolling_corr.append(corr)
    return np.array(rolling_corr)
```

```python
# 滑动相关性
window = 60
rolling_correlation = calculate_rolling_correlation(data['Temperature'], data['Electricity'], window)

# 对齐滑动窗口的时间轴
rolling_days = np.arange(len(rolling_correlation)) + window // 2

# 添加标题和图例
fig, ax2 = plt.subplots(figsize=(8, 6))

# fig.suptitle('Temperature and Electricity Usage with Seasonal Transitions', fontsize=16)
# ax1.legend(loc='upper left')
# ax2.legend(loc='upper right')
plt.title('Correlation Between Temperature and Electricity Usage', fontsize=16, pad=-1)

# 右轴：相关性
# fig.suptitle('Correlation Between Temperature and Electricity Usage ', fontsize=16)
ax2.plot(rolling_days, rolling_correlation, color='green', label='Rolling Correlation (30 days)', linewidth=2)
ax2.tick_params(axis='y')
# ax2.axhline(0, linewidth=1)
ax2.set_ylabel('Correlation')
# 自定义 x 轴的刻度标签
custom_ticks = ['A', 'B', 'C', 'D']
tickx = list(map(int, np.linspace(80, 300, len(rolling_days))))
plt.xticks( [rolling_days[i] for i in range(0, len(rolling_days), 30)], [tickx[i] for i in range(0, len(tickx), 30)])
# custom_ticks_positions = range(1, 21, 2)  # 每隔一个显示一个刻度
# custom_ticks_labels = [str(i) for i in custom_ticks_positions]
plt.savefig('elec_temp_corre.svg')
plt.tight_layout()
plt.show()
```

```python
# 滑动相关性

# 滑动相关性计算函数
def calculate_correlation(x, y, window):
    rolling_corr = []
    for i in range(len(x) - window + 1):
        corr = np.corrcoef(x[i:i+window], y)[0, 1]
        rolling_corr.append(corr)
    return np.array(rolling_corr)

window = 160
rolling_correlation = calculate_correlation(data['Temperature'], data['Temperature'][-window:], window)

# 对齐滑动窗口的时间轴
rolling_days = np.arange(len(rolling_correlation)) + window // 2

# 添加标题和图例
fig, ax2 = plt.subplots(figsize=(8, 6))

# fig.suptitle('Correlation With The Forecast Values', fontsize=16, pad=-1)
# ax1.legend(loc='upper left')
# ax2.legend(loc='upper right')
plt.title('Correlation With The Forecast Values', fontsize=16, pad=-1)

# 右轴：相关性
ax2.set_ylabel('Correlation')
ax2.plot(rolling_days, (rolling_correlation-rolling_correlation.min())/(rolling_correlation-rolling_correlation.min()).max(), color='green', label='Rolling Correlation (30 days)', linewidth=2)
# ax2.plot(rolling_days, rolling_correlation, color='green', label='Rolling Correlation (30 days)', linewidth=2)
tickx = list(map(int, np.linspace(80, 280, len(rolling_days))))
# ax2.set_xticks( [rolling_days[i] for i in range(0, len(rolling_days), 30)])
# ax2.set_xticklabels([tickx[i] for i in range(0, len(tickx), 30)])
plt.xticks( [rolling_days[i] for i in range(len(rolling_days))], tickx)

ax2.tick_params(axis='y')
# ax2.axhline(0, linewidth=1)

plt.savefig('future_correlation.svg')
plt.tight_layout()
plt.show()
```
